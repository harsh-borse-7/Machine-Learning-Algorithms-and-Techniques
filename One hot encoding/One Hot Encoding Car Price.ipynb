# Car price (One hot encoding)

### Importing necessary libraries

```python
import pandas as pd
from sklearn import linear_model
```

### Reading Excel File and making dataset

```python
df=pd.read_excel("One_hot_encoding_sheet_1-2.xlsx",sheet_name='Sheet1')
df
```

```python
df1=pd.get_dummies(df.city)
df1
```

```python
df2=pd.concat([df,df1],axis=1)
df2
```

### To Avoid Dummy variable trap

```python
df3_final=df2.drop(['city','C'],axis='columns')
df3_final
```

```python
reg=linear_model.LinearRegression()
reg.fit(df3_final[['area','A','B']],df3_final.price)
```

```python
import pickle 
with open ("one_hot_encoding_home.pkl",'wb') as f:
    pickle.dump(reg,f)
```

```python
with open("one_hot_encoding_home.pkl",'rb') as f:
    model=pickle.load(f)
model.predict([[3500,1,0],[35000,0,1],[35000,0,0]])

```
