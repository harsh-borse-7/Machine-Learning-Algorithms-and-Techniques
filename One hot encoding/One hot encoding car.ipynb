# Car Names (One Hot Encoding) 

### Importing necessary libraries

```python
import pandas as pd
from sklearn import linear_model
```

### Reading Excel File and making dataset

```python
df1=pd.read_excel('One_hot_encoding_sheet_1-2.xlsx',sheet_name="Sheet2")
df1
```

```python
df2=pd.get_dummies(df1.Car)
df2
```

```python
df3=pd.concat([df1,df2],axis='columns')
df3
```

### To Avoid Dummy variable trap

```python
df=df3.drop(['Car','Audi'],axis=1)
df                
```

### Building Model

```python
reg=linear_model.LinearRegression()
reg.fit(df3[['mileage','age','Audi','BMW']],df3.price)
```

```python
import pickle
with open("one_hot_encoding_car.pkl",'wb') as f:
    pickle.dump(reg,f)
```

```python
with open("one_hot_encoding_car.pkl",'rb') as f:
    model=pickle.load(f)
```

```python
model.predict([[85000,5,0,1],[85000,5,1,0],[85000,5,0,0]])
```
