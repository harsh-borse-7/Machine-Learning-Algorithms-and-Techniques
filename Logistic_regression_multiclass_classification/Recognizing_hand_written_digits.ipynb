# Recognizing hand written digits

### Importing necessary libraries

```python
from sklearn.datasets import load_digits
from sklearn import linear_model
import matplotlib.pyplot as plt
%matplotlib  inline
```

### Loading Dataset

```python
digits_df=load_digits()
```

```python
dir(digits_df)
```

### Spliting data for training and testing

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(digits_df.data,digits_df.target,test_size=0.2)
```

### Building model

```python
model=linear_model.LogisticRegression()
model.fit(X_train,y_train)
```

### Predicting Values of test dataset

```python
y_pred = model.predict(X_test) 
```

### Scores

```python
model.score(X_test,y_test)
```

```python
plt.gray()
for i in range(5,10):
    plt.matshow(digits_df.images[i])
```

```python
model.predict(digits_df.data[5:10])
```

### Confusion matrix

```python
from sklearn.metrics import confusion_matrix
```

```python
y_predicted=model.predict(X_test)
cm=confusion_matrix(y_predicted,y_test)
cm
```

```python
import seaborn as sns
sns.heatmap(cm, annot=True)
```

### Classification Report

```python
from sklearn.metrics import classification_report
cr=classification_report(y_test,y_pred)
print(cr)
```
