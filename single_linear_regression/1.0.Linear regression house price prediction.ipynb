# predicting house price using linear regression

### importing essential libraries 


```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

### creating dataframe from excel file

```python
df=pd.read_excel("ML_1_linear_regression.xlsx")
```

```python
print(df)
```

### Finding correlation matrix

```python
df.corr()
```

### Making linear regression model

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

### Finding slope and intercept of line of best fit (Regression line)

```python
m=reg.coef_
c=reg.intercept_
```

### ploting using matplotlib

```python
plt.plot(df.area,m*df.area+c)
plt.scatter(df.area,df.price,color='red',marker="*")
```

```python
reg.predict([[3300],[2500],[3800]])
```


### Ploting using Seaborn


```python
import seaborn as sns
sns.regplot(df.area,df.price,ci=None)
```

### Storing Model in Pickle file

```python
import pickle
```

```python
with open('house_model.pickle','wb') as f:
    pickle.dump(reg,f)
```

```python
with open('house_model.pickle','rb') as f:
    model=pickle.load(f)
```

```python
model.predict([[5000]])
```
